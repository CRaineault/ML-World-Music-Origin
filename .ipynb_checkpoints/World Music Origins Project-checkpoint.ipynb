# World Music Origin Project

## Abstract

We took the problem of identifying the location of origin of music and attempted to represent it as a neural network, classifying the location of origin. While our results were marginally successful, we determined that standard neural network architecture is an insufficiently accurate method of categorizing the features of music.


## Introduction

Music is common to all people around the globe. Debates range as to how to define music, but most musicologists agree that any structured sound that is intended to be listened to for its own sake, and that is intended to be perceived as music, counts as music. That being said, every culture around the world has patterns for creating music as a part of that culture, including instruments, melodic and harmonic devices, timbres, and rhythmic patterns. Humans can, with some training, identify the elements of a given piece of music purely from auditory input, and with that information can locate the country of origin for a piece of world music. In current Western music nomenclature, “world music” refers to any piece of music that doesn’t fall into traditional notions of Western music- in other words, Roots magazine defines it as "local music from out there". This nebulous definition then means that the term is often derided as being over-broad, but it’s a useful term for attracting people who wish to have a broad musical palette. This poses a problem for musical libraries of world music (such as that of the Sonic Planet program on the radio station WICB)- if the music is initially uncategorized, it can be challenging to find more information about the music until its location of origin can be determined. While humans can do this aurally, expediting the process with machine learning would help to sift through large batches of music. We hypothesize that the aural characteristics of music can be analyzed by a neural network to locate the origin of individual pieces of music.

## Related work

The collection of data has already been done by Zhou et. al. This team identified previous work on identifying the sonic aspects of music (including spectral centroids), and using these aspects to recommender systems, web searches, and social tags. They were less successful in identifying computational ethnomusicology, though Liu et. al. demonstrated that music analysis techniques could be applied to non-western music. Zhou et. al. also did their own previous work in analyzing this dataset, and approached it from a regression perspective where location is compared by distance to the source. This method was chosen for its musicological applications, highlighting cultural exchange across different cultures. The model they chose was a random forest regression model, which yielded an average distance error of just over 3000 km.

## Methods

The final model chosen used the chromatic features analyzed by Zhou et. al. as the input data. 1,142 songs had been selected from 33 countries and analyzed through MARSYAS’s default single-vector format. While the data given was presented as the latitude and longitude of the capital of the country of origin, we converted this data into a one-hot encoded matrix with each category representing the country. We chose this method for its commercial applications, as these mostly use country of origin as the sub-categorization. This neural network was compared to models of simple logistic regression and assuming the most common country of origin.

```python
#imports, add more as needed
import os

import numpy as np

import sklearn as skl

from sklearn import model_selection

from sklearn import datasets

from sklearn import linear_model

from sklearn import metrics

import tensorflow as tf

from matplotlib import pyplot as plt

from geopy.geocoders import Nominatim

from tensorflow.keras import Sequential

from tensorflow.keras.layers import Dense

from tensorflow.keras.losses import CategoricalCrossentropy

from tensorflow.keras.activations import softmax

from tensorflow.keras.optimizers import Adam

from random import randint
```

## Train, CV, Test split

```python
#splitting data into train, cv, test

split = skl.model_selection.train_test_split(X, y)

train_x = split[0]
train_y = split[2]

cv_split = skl.model_selection.train_test_split(split[1], split[3])

cv_x = cv_split[0]

cv_y = cv_split[2]

test_x = cv_split[1]

test_y = cv_split[3]
```

## Logistic Regression Model

```python
#One v. all logitistic regression

regression_model = Sequential([
    Dense(units=33, activation ='linear')
    
])

regression_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss=CategoricalCrossentropy(from_logits=True))

print(train_x.shape)

print(train_y.shape)

regression_model.fit(train_x, train_y, epochs=2200)

logits = regression_model(train_x)

prediction = tf.nn.softmax(logits)
```

```python
p = np.argmax(prediction, axis=1)
predict_matrix = np.zeros(train_y.shape)
for r in range(prediction.shape[0]):
    predict_matrix[r, p[r]] = 1;

    
print(predict_matrix)
    
```

## Neural Network Model

```python
#neural network model

nn_model = Sequential([
    Dense(units=128, activation = 'relu'),
    Dense(units=64, activation = 'relu'),
    Dense(units=33, activation ='relu')
    
])


lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    1e-3,
    decay_steps=10000,
    decay_rate=0.55,
    staircase=False)

nn_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=lr_schedule),loss=CategoricalCrossentropy(from_logits=True))

print(train_x.shape)

print(train_y.shape)

nn_model.fit(train_x, train_y, epochs=2000)

logits = nn_model(train_x)

nn_prediction = tf.nn.softmax(logits)
```

```python
p = np.argmax(nn_prediction, axis=1)
nn_predict_matrix = np.zeros(train_y.shape)
for r in range(nn_prediction.shape[0]):
    nn_predict_matrix[r, p[r]] = 1;

    
print(nn_predict_matrix)
    
```

## Accuracy Testing Functions

```python
def testAccuracy(result):
    counter = 0
    for i in range(result.shape[0]):
        for n in range(result.shape[1]):
            if result[i, n] == 1 and result[i,n] == train_y[i, n]:
                counter += 1
                
    accuracy = (counter/result.shape[0]) * 100
    
    return accuracy
```

```python
def testAccuracyCV(result):
    counter = 0
    for i in range(result.shape[0]):
        for n in range(result.shape[1]):
            if result[i, n] == 1 and result[i,n] == cv_y[i, n]:
                counter += 1
                
    accuracy = (counter/result.shape[0]) * 100
    
    return accuracy
    
```

```python
def testAccuracyTest(result):
    counter = 0
    for i in range(result.shape[0]):
        for n in range(result.shape[1]):
            if result[i, n] == 1 and result[i,n] == test_y[i, n]:
                counter += 1
                
    accuracy = (counter/result.shape[0]) * 100
    
    return accuracy
```

## Random Choice Model

```python
#random model for accuracy testing

random_matrix = np.zeros(train_y.shape)

for r in range(random_matrix.shape[0]):
    label = randint(0, 32)
    random_matrix[r, label] = 1
    
print(random_matrix)
    

```

## Preparing Result Lists for Confusion Matrices

```python
train_results = []

for r in range(train_y.shape[0]):
    num_label = np.where(train_y[r] == 1)[0]
    train_results.append(num_label[0])
```

```python
predict_results = []
predict_countries = countries.copy()
for r in range(predict_matrix.shape[0]):
    num_label = np.where(predict_matrix[r] == 1)[0]
    predict_results.append(num_label[0])

for i in range(33):
    if i not in predict_results and i not in train_results:
        country = countries[i]
        predict_countries.remove(country)
```

```python
nn_predict_results = []
nn_predict_countries = countries.copy()

for r in range(nn_predict_matrix.shape[0]):
    num_label = np.where(nn_predict_matrix[r] == 1)[0]
    nn_predict_results.append(num_label[0])
    
for i in range(33):
    if i not in nn_predict_results and i not in train_results:
        country = countries[i]
        nn_predict_countries.remove(country)
```

```python
test_results = []


for r in range(test_y.shape[0]):
    num_label = np.where(test_y[r] == 1)[0]
    test_results.append(num_label[0])
```

```python
test_predict_results = []
test_predict_countries = countries.copy()
for r in range(test_predict_matrix.shape[0]):
    num_label = np.where(test_predict_matrix[r] == 1)[0]
    test_predict_results.append(num_label[0])
    
for i in range(33):
    if i not in test_predict_results and i not in test_results:
        country = countries[i]
        test_predict_countries.remove(country)
```

```python
nn_test_predict_results = []
nn_test_predict_countries = countries.copy()

for r in range(nn_test_predict_matrix.shape[0]):
    num_label = np.where(nn_test_predict_matrix[r] == 1)[0]
    nn_test_predict_results.append(num_label[0])
    
for i in range(33):
    if i not in nn_test_predict_results and i not in test_results:
        country = countries[i]
        nn_test_predict_countries.remove(country)
```

## Creating Confusion Matrices

```python
confusion_matrix = skl.metrics.confusion_matrix(train_results, predict_results)
confusion_matrix2 = skl.metrics.confusion_matrix(train_results, nn_predict_results)
confusion_matrix3 = skl.metrics.confusion_matrix(test_results, test_predict_results)
confusion_matrix4 = skl.metrics.confusion_matrix(test_results, nn_test_predict_results)
```

## Plotting Confusion Matrices

```python
def plot_confusion_matrix(confusion_matrix, subtitle, country_list):
    fig, ax = plt.subplots(figsize=(7.5, 7.5))
    ax.matshow(confusion_matrix, cmap=plt.cm.Blues, alpha=0.3)
    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(start, end, 1.0))
    ax.set_xticklabels(country_list, fontsize=8)
    ax.yaxis.set_ticks(np.arange(start, end, 1.0))
    ax.set_yticklabels(country_list, fontsize=8)
    for i in range(confusion_matrix.shape[0]):
        for j in range(confusion_matrix.shape[1]):
            ax.text(x=j, y=i,s=confusion_matrix[i, j], va='center', ha='center', size='small')

    plt.xlabel('Predictions', fontsize=18)
    plt.ylabel('Actuals', fontsize=18)
    plt.title(subtitle + ' Confusion Matrix', fontsize=12)
    plt.show()
```

```python
plot_confusion_matrix(confusion_matrix, "Logistic Regression Train Set", predict_countries)
plot_confusion_matrix(confusion_matrix2, "NN Train Set", nn_predict_countries)
plot_confusion_matrix(confusion_matrix3, "Logistic Regression Test Set", test_predict_countries)
plot_confusion_matrix(confusion_matrix4, "NN Test Set", nn_test_predict_countries)
```

## Conclusion

There are multiplefactors that may have contributed to the poor success of the models in this project. One factor is our consistent trouble with the neural network model suddenly exploding with high loss. While the learning rate scheduler mitigates this problem, it seems for some reason the neural network cannot reach much of a lower cost and thus higher accuracy without facing this issue.

However, since both models did roughly the same on the CV and test data, it is likely fixing the neural network would not fix the problem. The larger issue seems to be the fact that using extracted numerical features and chromatic data with neural networks results in a combination of old and new ML technologies that do not work well together in making accurate predictions. What instead may solve the world music origin prediction problem more satisfactorily is a more robust dataset compatible with convolutional neural networks and the neural network revolution of ML, which the provided MARSYAS data does not. Future work on this topic would likely benefit from a different, more fitting dataset (if one exists) that could be used to accomplish this task.
